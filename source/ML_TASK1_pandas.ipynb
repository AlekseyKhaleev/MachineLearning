```python
from __future__ import (absolute_import, division, print_function, unicode_literals)
import warnings
warnings.simplefilter('ignore')
%pylab inline

from pylab import rcParams
rcParams['figure.figsize'] = 8, 5
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('Video_Games_Sales_as_at_22_Dec_2016.csv')
df.info()
```

```python
df = df.dropna()
print(df.shape)
```

```python
useful_cols = ['Name', 'Platform', 'Year_of_Release', 'Genre', 'Global_Sales', 'Critic_Score', 'Critic_Count', 'User_Score', 'User_Count', 'Rating']
df[useful_cols].head()
```

```python
sales_df = df[[x for x in df.columns if 'Sales' in x] + ['Year_of_Release']]
sales_df.head()
```

```python
df1 = sales_df.groupby('Year_of_Release').sum()
print(df1)
```

```python
x = sales_df.groupby('Year_of_Release').sum()
x = x.apply(pd.to_numeric, errors='coerce')
x.plot()
```

```python
sales_df.groupby('Year_of_Release').sum().plot(kind='bar', rot=45)
```

```python
cols = ['Global_Sales', 'Critic_Score', 'Critic_Count']
sns_plot = sns.pairplot(df[cols])
```

```python
top_platforms = df.Platform.value_counts().sort_values(ascending = False).head(5).index.values
sns.boxplot(y="Platform", x="Critic_Score", data=df[df.Platform.isin(top_platforms)], orient="h")
```

```python
sns.heatmap(df[cols].corr(), annot=True, fmt=".1f", linewidths=.5)
```
